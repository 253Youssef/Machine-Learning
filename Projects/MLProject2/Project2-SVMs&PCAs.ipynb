# Task 1

## Dataset

```python
# #############################################################################
# Download the data, if not already on disk and load it as numpy arrays

lfw_people = fetch_lfw_people(min_faces_per_person=70, resize=0.4)

# introspect the images arrays to find the shapes (for plotting)
n_samples, h, w = lfw_people.images.shape

# for machine learning we use the 2 data directly (as relative pixel positions info is ignored by this model)
X = lfw_people.data
n_features = X.shape[1]

# the label to predict is the id of the person
y = lfw_people.target
target_names = lfw_people.target_names
n_classes = target_names.shape[0]

print("Total dataset size:")
print("n_samples: %d" % n_samples)
print("n_features: %d" % n_features)
print("n_classes: %d" % n_classes)

# #############################################################################
# Split into a training set and a test set using a stratified k fold

# split into a training and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

## Functions

```python
def plot_gallery(images, titles, h, w, n_row=3, n_col=4):
    """Helper function to plot a gallery of portraits"""
    plt.figure(figsize=(1.8 * n_col, 2.4 * n_row))
    plt.subplots_adjust(bottom=0, left=.01, right=.99, top=.90, hspace=.35)
    for i in range(n_row * n_col):
        plt.subplot(n_row, n_col, i + 1)
        plt.imshow(images[i].reshape((h, w)), cmap=plt.cm.gray)
        plt.title(titles[i], size=12)
        plt.xticks(())
        plt.yticks(())


# plot the result of the prediction on a portion of the test set

def title(y_pred, y_test, target_names, i):
    pred_name = target_names[y_pred[i]].rsplit(' ', 1)[-1]
    true_name = target_names[y_test[i]].rsplit(' ', 1)[-1]
    return 'predicted: %s\ntrue:      %s' % (pred_name, true_name)
```

## Original Eigenfaces Tutorial

```python
# #############################################################################
# Compute a PCA (eigenfaces) on the face dataset (treated as unlabeled dataset): unsupervised feature extraction / dimensionality reduction
n_components = 150

print("Extracting the top %d eigenfaces from %d faces"% (n_components, X_train.shape[0]))
t0 = time()
pca = PCA(n_components=n_components, svd_solver='randomized',whiten=True).fit(X_train)
print("done in %0.3fs" % (time() - t0))

eigenfaces = pca.components_.reshape((n_components, h, w))

print("Projecting the input data on the eigenfaces orthonormal basis")
t0 = time()
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
print("done in %0.3fs" % (time() - t0))

# #############################################################################
# Train a SVM classification model

print("Fitting the classifier to the training set")
t0 = time()
param_grid = {'C': [1e3, 5e3, 1e4, 5e4, 1e5],'gamma': [0.0001, 0.0005, 0.001, 0.005, 0.01, 0.1], }
clf = GridSearchCV(SVC(kernel='rbf', class_weight='balanced'),param_grid, cv=5)
clf = clf.fit(X_train_pca, y_train)
print("done in %0.3fs" % (time() - t0))
print("Best estimator found by grid search:")
print(clf.best_estimator_)

# #############################################################################
# Quantitative evaluation of the model quality on the test set

print("Predicting people's names on the test set")
t0 = time()
y_pred = clf.predict(X_test_pca)
print("done in %0.3fs" % (time() - t0))

print(classification_report(y_test, y_pred, target_names=target_names))
print(confusion_matrix(y_test, y_pred, labels=range(n_classes)))

# #############################################################################
# Qualitative evaluation of the predictions using matplotlib

prediction_titles = [title(y_pred, y_test, target_names, i) for i in range(y_pred.shape[0])]

plot_gallery(X_test, prediction_titles, h, w)

# plot the gallery of the most significative eigenfaces

eigenface_titles = ["eigenface %d" % i for i in range(eigenfaces.shape[0])]
plot_gallery(eigenfaces, eigenface_titles, h, w)

plt.show()
```

## Task 1.3: Various Number of Principal Components

```python
# #############################################################################
# Compute a PCA (eigenfaces) on the face dataset (treated as unlabeled dataset): unsupervised feature extraction / dimensionality reduction
principal_components = [50, 150, 190, 300, 700]
accuracy_scores = []
c_values = []

for number in principal_components:
  
  n_components = number

  print("Extracting the top %d eigenfaces from %d faces"% (n_components, X_train.shape[0]))
  t0 = time()
  pca = PCA(n_components=n_components, svd_solver='randomized',whiten=True).fit(X_train)

  eigenfaces = pca.components_.reshape((n_components, h, w))

  print("Projecting the input data on the eigenfaces orthonormal basis")
  t0 = time()
  X_train_pca = pca.transform(X_train)
  X_test_pca = pca.transform(X_test)

  # #############################################################################
  # Train a SVM classification model

  t0 = time()
  param_grid = {'C': [1, 10, 100, 1e3],'gamma': [0.0001, 0.0005, 0.001, 0.005, 0.01, 0.1], }
  clf = GridSearchCV(SVC(kernel='rbf', class_weight='balanced'),param_grid, cv=5)
  clf = clf.fit(X_train_pca, y_train)
  c_values.append(clf.best_params_['C'])

  # #############################################################################
  # Quantitative evaluation of the model quality on the test set

  t0 = time()
  y_pred = clf.predict(X_test_pca)
  
  accuracy_scores.append(clf.score(X_test_pca, y_test))
```

## Task 1.4: Performance Report

As shown in the graph of the accuracy scores above we can clearly see that accuracy score increase from using 50 principal components to 150, then continues to decrease after that. This shows that the optimal number of principal components in this case is 150 principal components.

This observation makes sense since we have a high score due to using the 150 principal compoenent with the highest variance, in addition to the fact that SVMs use a quadratic solver that works best with handling smaller datasets. So it makes perfect sense that giving the SVM a smaller number of principal components yet of great relvance to the SVM classifier which works best with smaller datasets would yield the optimal solution in this case.

### Task 2.1.c: Generalization and Overfitting

It is evident from the graph provided in the previous section that the training score increases as the number of principal components used increases until 110 principal components where the training score becomes 1.00, while the accuracy scores keep flactuating around 0.7 in the begining before 110 principal components, then it decreases drastically to around 0.1.



Putting these observations into perspective, we can clearly see that the optimal number of principal components to use is 105 where it has a relatively high accuracy score while mainting a training score of around 0.98 which shows that the model does not overfit yet and can be used generally.



An important observation about the scores relative to the number of principal components used is that the more principal component we use, the more we overfit our model and prevent it from working generally, as it is evident by the drastic change in the accuracy score between 110 and 115 principal components and how that accuracy score continues on being low, which makes sense when we observe the same number of principal components where we see that the training score becomes 1.0 from 105 principal components to 110, where the model is overfitted on the data it trains on and is no longer capable of being used for general datapoints.

### MeanShift Algorithm Explanation

MeanShift algorithm is a hierarchical clustering algorithm that works by assigning points iteratively by the use of a parameter called bandwidth that forms some sort of a raduis around a given datapoint and finds the mean of all datapoints in this given circle, then the algorithm iteratively does the same process again on the previosuly formed mean point until it converges to 1 point assigning all of the datapoints involved into 1 cluster.

To provide more technical information on the algorithm, Meanshift is basically a clustering algorithm that assigns the datapoints to the clusters iteratively by shifting points towards the mode which can be understood as the highest density of datapoints (in the region, in the context of the Meanshift). This approach heavily depends on the Kernel Density Estimation (KDE) which is a method to estimate the underlying probability of a given dataset.

The advantage of MeanShift algorithm over other cclustering techniques is that it does not require a given parameter for the number of clusters since it determines that on its own. Regardess of the relatively wrong number of datapoints per cluster, we can see that it figured out 13 clusters on its own out of the 14 actual clusters.

## AgglomerativeClustering

```python
clustering = AgglomerativeClustering(n_clusters=14, linkage='average').fit(np.concatenate([X_train, X_test]))

predicted_labels = ((clustering.labels_)).tolist()
actual_labels = np.concatenate([y_train, y_test]).astype('int')

actual_labels = actual_labels.tolist()
                             
print('Predicted Labels:', predicted_labels)
      
print('Actual Labels:', actual_labels)

# Create dictionary t assign prediction labels to atual labels since they are assigned differently with a defualt value of -1 (Used for later check)
# Example: Actual Label 1 might be equivalent to Predicted Label 10
dictionary = defaultdict(lambda: -1)

# Index to loop over all labels
index = len(actual_labels) - 1

# Loop over all labels of both lists
while index != 0:
  # If the value for key of actual_labels[index] is -1, then that means that the predicted label mapping for the actual label was not set yet
  if (dictionary[actual_labels[index]] == -1):
    # Set the actual label mapping to the predicted label
    dictionary[actual_labels[index]] = predicted_labels[index]
  index = index - 1
  
# Number of correct mappings
correct = 0

index = len(actual_labels) - 1

while index != 0:
  if (dictionary[actual_labels[index]] == predicted_labels[index]):
    correct = correct + 1
  index = index - 1
  
print('Correct Mappings of Labels:', correct / len(actual_labels))
```

### AgglomerativeClustering Explanation

AgglomerativeClustering is a hierarchical clustering algorithm that works by a bottom up approach, where datapoints are given initial clusters and clusters merge if a certain criterion is present in both clusters according to a given merging strategy. n_clusters parameter sets the number of clusters the clustering algorithm will attempt to find. That is, clusters are successively merged until there are only n_clusters remaining.

AgglomerativeClustering forms a dendogram for the clustered datapoints.
