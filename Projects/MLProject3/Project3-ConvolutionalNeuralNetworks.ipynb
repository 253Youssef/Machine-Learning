# Libraries

```python
from keras import backend as K
print(K.tensorflow_backend._get_available_gpus())

import os as os
from os import listdir
from os.path import isfile, join

import numpy as np
import matplotlib.pyplot as plt
import keras
from keras.utils import to_categorical
from keras.callbacks import ModelCheckpoint
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, Dropout, Flatten
from keras.layers.normalization import BatchNormalization
from sklearn.model_selection import train_test_split
```

# Model Creation

```python
# Specify the input shape to the first convolutional layer
input_shape = (120, 160, 3)
nClasses = 2
def createModel():
    model = Sequential()
    # a convolution layer of 32 features of size 3x3 with relu activation and zero padding
    model.add(Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=input_shape))
    # a convolution layer of 32 features of size 3x3 with relu activation
    model.add(Conv2D(32, (3, 3), activation='relu'))
    # a batch normalization layer
    model.add(BatchNormalization())
    # maxpooling layer of filter size 2x2
    model.add(MaxPooling2D(pool_size=(2, 2)))
    
    # a convolution layer of 64 features of size 3x3 with relu activation and zero padding
    model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
    # a convolution layer of 64 features of size 3x3 with relu activation
    model.add(Conv2D(64, (3, 3), activation='relu'))
    # a batch normalization layer
    model.add(BatchNormalization())
    # maxpooling layer of filter size 2x2
    model.add(MaxPooling2D(pool_size=(2, 2)))
    
    # a convolution layer of 64 features of size 3x3 with relu activation and zero padding
    model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
    # a convolution layer of 64 features of size 3x3 with relu activation
    model.add(Conv2D(64, (3, 3), activation='relu'))
    # a batch normalization layer
    model.add(BatchNormalization())
    # maxpooling layer of filter size 2x2
    model.add(MaxPooling2D(pool_size=(2, 2)))
    # a dropout layer of 50%
    model.add(Dropout(0.5))
    
    # flatten the output of the previous layer
    model.add(Flatten())
    # add a dense layer that outputs 512 units and apply relu activation
    model.add(Dense(512, activation='relu'))
    # a dropout layer of 50%
    model.add(Dropout(0.5))
    # add a dense layer with a softmax activation to classify the images
    model.add(Dense(nClasses, activation='softmax'))
     
    return model
```

# Model Training

```python
# Create model using the provided createModel() function and load the weights 
model1 = createModel()

# Define a learning rate of 0.0001 to have a slow but consistent training result
learning_rate = 0.0001

# Use sgd optimizer with the specified learning rate
optimizer = keras.optimizers.SGD(lr = learning_rate)

# Load the provided weights file
model1.load_weights('weights.hdf5')

# Compile the model using the defined optimizer, a loss function of crossentropy, and a metric of accuracy
model1.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Define batch size and number of epochs
epochs = 30

# Define a batch size for the training dataset of 32 to have a balance between optimization and quality training
train_batch_size = 32
# Define steps per epoch to be the number of total training images over the batch size as it is the batch iterations required in the training before a training epoch has finished.
steps_per_epoch = int(3872*0.8 / train_batch_size)

# Define a batch size for the validation dataset of 16 since the validation dataset is less than the training dataset
validation_batch_size  = 16
validation_steps = int(3872*0.2 / validation_batch_size)

# Define a batch size for the testing dataset of 16 since the testing dataset is less than the training dataset
test_batch_size = 16
test_steps = int(860 / test_batch_size)


train_datagen = ImageDataGenerator(
        rescale = 1./255,
        samplewise_center = True,
        samplewise_std_normalization=True, # divide each input by its std.
        validation_split = 0.2)   

# Create a checkpoint to save the best weights after the training
checkpointer = ModelCheckpoint(filepath='best_weights.hdf5', verbose=1, save_best_only=True)
callbacks_list = [checkpointer]

# Initilaise train_generator
train_generator = train_datagen.flow_from_directory(
        "./train",
        target_size=(120, 160),
        batch_size=train_batch_size,
        class_mode='binary',
        subset="training"
    )

# Initilaise validation_generator
validation_generator = train_datagen.flow_from_directory(
        "./train",
        target_size=(120, 160),
        batch_size=validation_batch_size,
        class_mode='binary',
        subset="validation"
    )

# Initilaise test_generator
test_generator = train_datagen.flow_from_directory(
        "./test",
        target_size=(120, 160),
        batch_size=test_batch_size,
        class_mode='binary'
    )

# Fit the model and store its history
history1 = model1.fit_generator(
    train_generator,
    steps_per_epoch = steps_per_epoch,
    epochs = epochs,
    validation_data = validation_generator,
    validation_steps = validation_steps,
    callbacks = callbacks_list
)

# Evaluate the model using the test_generator
evaluation = model1.evaluate_generator(generator = test_generator, steps = test_steps)
names = model1.metrics_names

# Display metrics and their corresponding scores
for name, score in zip(names, evaluation):
    print(name, score)
```

### Choice of batch_size, steps_per_epoch, and validation_steps

We use a batch size of 32 in order to speed up the process of training, but not a large value of a batch size as not to compromise the training process. Since an epoch is 1 pass over the entire training data throughout the network, the steps per epoch used is the total number of images over tha batch size as this is the number of batch iterations required in the training before a training epoch has finished. Same reasoning is used for the validation steps.

# Plot the train and validation loss and accuracy curves

```python
# Loss Curves
plt.figure(figsize=[8,6])
plt.plot(history1.history['loss'],'r',linewidth=3.0)
plt.plot(history1.history['val_loss'],'b',linewidth=3.0)
plt.legend(['Training loss', 'Validation Loss'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Loss',fontsize=16)
plt.title('Loss Curves',fontsize=16)
 
# Accuracy Curves
plt.figure(figsize=[8,6])
plt.plot(history1.history['acc'],'r',linewidth=3.0)
plt.plot(history1.history['val_acc'],'b',linewidth=3.0)
plt.legend(['Training Accuracy', 'Validation Accuracy'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Accuracy',fontsize=16)
plt.title('Accuracy Curves',fontsize=16)
```
